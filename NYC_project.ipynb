## 2.3 Analyze the date column, and remove entries that have an incorrect timeline


```python
pd.to_datetime(data1['Created Date'],format='%Y-%m-%d',errors='coerce').isna().sum()
```

```python
pd.to_datetime(data1['Closed Date'],format='%Y-%m-%d',errors='coerce').isna().sum()
```

### 2.3.1 Calculate the time elapsed in closed and creation date


```python
# checking the datatype of time

display(data1['Closed Date'].dtype,data1['Created Date'].dtype)
```

```python
# as the time is in object format we need to change the datatype

data1['Closed Date'] = pd.to_datetime(data1['Closed Date'])
```

```python
data1['Created Date'] = pd.to_datetime(data1['Created Date'])
```

```python
data1.info()
```

```python
# now calculating the time elapsed 
time_elapsed = data1['Closed Date'] - data1['Created Date']
time_elapsed.head(5)
```

### 2.3.2 Convert the calculated date to seconds to get a better representation


```python
data1['time_elapsed_seconds'] = time_elapsed.dt.total_seconds()
```

### 2.3.5 Impute the NA value with Unknown City

```python
data1['City']=data1['City'].replace(np.nan,'Unknown City')

# data1['City'].fillna('Unknown City')  can be used in place of replace
```

```python
data1['City'].isna().sum()
```

### 2.3.6 Draw a frequency plot for the complaints in each city

```python
complaint_count = data1.City.value_counts()
complaint_count
```

```python
plt.figure(figsize=(8,12))
plt.barh(y=complaint_count.index,width=complaint_count.values,color='y')
plt.xticks(rotation=45)
plt.xlabel('Complaints Count')
plt.ylabel('City')
plt.title('City VS Complaints')
plt.show()
```

### 2.3.7 Create a scatter and hexbin plot of the concentration of complaints across Brooklyn

```python
brooklyn = data1.query("City=='BROOKLYN'")
brooklyn
```

```python
brooklyn.plot.scatter(x='Longitude', y='Latitude', title = 'Complaints Concentration in Brooklyn', figsize = (10, 8))
plt.show()
```

```python
brooklyn.plot.hexbin(x='Longitude', y='Latitude', gridsize=40,
    colormap = 'jet', mincnt=1, title = 'Complaints Concentration in Brooklyn', figsize = (10, 8))
plt.show()
```

## 3. Find major types of complaints:


### 3.1 Plot a bar graph to show the types of complaints

```python
complaints = data1['Complaint Type'].value_counts()
complaints.plot.bar(figsize=(6,6))
plt.xlabel('Complaints')
plt.ylabel('Frequency of Complaints')
plt.title('Types of Complaints')
plt.show()
```

### 3.2 Check the frequency of various types of complaints for New York City

```python
new_york_city = data1[data1.City == 'NEW YORK']
new_york_city['Complaint Type'].value_counts().plot(kind='bar')
```

### 3.3 Find the top 10 complaint types

```python
top_10_complaints = data1['Complaint Type'].value_counts().head(10)
top_10_complaints.plot.bar(figsize=(6,6))
plt.xlabel('Complaints')
plt.ylabel('Frequency of Complaints')
plt.title('Top 10 Complaints')
plt.show()
```

### 3.4 Display the various types of complaints in each city


```python
city_complaint = data1.groupby(['City'])[['Complaint Type']]
city_complaint.value_counts()
```

### 3.5 Create a DataFrame, df_new, which contains cities as columns and complaint types in rows

```python
city = []
complaint_type = []
for i,j in city_complaint:
    city.append(i)
    complaint_type.append(set(j['Complaint Type'].values))
```

```python
citywise_complaint_type = pd.DataFrame(data=complaint_type,index=city).T
citywise_complaint_type
```
