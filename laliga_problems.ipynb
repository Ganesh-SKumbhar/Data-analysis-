### 1. Read data set and replace dashes(-)  with 0

```python
laliga = laliga.replace('-',0)
```

### Checking the data types

```python
laliga.dtypes
```

```python
for column in laliga.columns[2:-3]:
    if laliga[column].dtype == 'object':
        laliga[column] = laliga[column].astype('int')
laliga.dtypes
```

## 3. Print the list of teams which came Top 5 in term of points

## 4. Write a function with name "Goal_diff_count" which should return all teams with their Goal differences.

```python
def Goal_diff_count():
    return laliga.GoalsFor - laliga.GoalsAgainst

laliga['GoalDifference'] = Goal_diff_count()
laliga[['Team','GoalDifference']]
```

### Using the same function, find the team which hav maximum and minimum goal difference

```python
# Finding maximum difference

laliga[['Team','GoalDifference']].head(1)
```

```python
# finding minimun difference

laliga[['Team','GoalDifference']].tail(1)
```

## 5. Create a new column name 'winning percent'  and append it to dataset


```python
laliga['WinningPercent'] = (laliga.GamesWon/laliga.GamesPlayed).round(2)*100
```

```python
laliga['WinningPercent']
```

```python
# here in abouve output we can see that row 60 : percentage is NaN

laliga.WinningPercent.isna().sum()
```

### 1 entry is NaN so we need to replace it by 0

```python
laliga.WinningPercent = laliga.WinningPercent.replace(np.nan,0)
```

### Print the top 5 teams which has the highest Winning percentage


```python
laliga.sort_values('WinningPercent', ascending=False).head(5)[['Team','WinningPercent']]
```
